<br/>
<div class="alert alert-info">

Заполнение и обработка пропусков

- В столбце `name` пропуски заменим на значение Unknown.

- В столбце `genre` пропуски заменим на значение Unknown.

- Пропуски в столбце с годом выпуска не будут обрабатываться, поскольку они пропадут после фильтрации данных до нужного среза.

- Пропуски в столбцах по продажам игр в Европе и Японии заменим на среднее значение по продажам в зависимости от платформы и года выхода игры.

- Пропуски в оценках и рейтинге останутся, поскольку такие данные сложно предположить.
</div>

```python
# заполнение пропусков в столбце name
df_copy['name'] = df_copy['name'].fillna('Unknown')
```

```python
# заполнение пропусков в столбце genre
df_copy['genre'] = df_copy['genre'].fillna('Unknown')
```

```python
# заполнение пропусков в данных по количеству проданных игр в Европе и Японии
# функции для подбора среднего значения в зависимости от платформы и года выпуска

def eu_sales(row):
    if pd.isna(row['eu_sales']):
        eu_group = df_copy[(df_copy['platform'] == row['platform']) 
                           & (df_copy['year_of_release'] == row['year_of_release'])]
        return eu_group['eu_sales'].mean()
    else:
        return row['eu_sales']

def jp_sales(row):
    if pd.isna(row['jp_sales']):
        jp_group = df_copy[(df_copy['platform'] == row['platform']) 
                           & (df_copy['year_of_release'] == row['year_of_release'])]
        return jp_group['jp_sales'].mean()
    else:
        return row['jp_sales']
```

```python
# применение функций для замены пропусков
df_copy['eu_sales'] = df_copy.apply(eu_sales, axis=1)
df_copy['jp_sales'] = df_copy.apply(jp_sales, axis=1)
```

- В данных присутствуют явные дубликаты. 241 строку из 169566 нужно удвлить, чтобы убрать явные дубликаты. Убираем явные дубликаты с помощью метода `drop_duplicates()`. Сохраняем новый датафрейм без дубликатов, оставляем только первую строку из дублирующихся.

```python
# сортируем датафрейм по всем столбцам
df_sorted = df_copy.sort_values(by=list(df_copy.columns))

# удаляем дубликаты
df_no_duplicates = df_sorted.drop_duplicates()
```

<br/>
<div class="alert alert-info">

- Часть строк в процессе предобработки была удалена. Чтобы вывести их количество в абсолютном и относительном значениях необходимо сравнить количество строк в датафрейме до обработки (`df`) и количество строк в датафрейме после обработки (`df_no_duplicates`)
</div>

```python
# количество строк в исходном датафрейме
initial_row_count = df.shape[0]

# количество строк в датафрейме после обработки
final_row_count = df_no_duplicates.shape[0]

# cчитаем сколько строк было удалено
deleted_rows = initial_row_count - final_row_count

# cчитаем относительное количество удаленных строк
deleted_share = round(deleted_rows / initial_row_count, 3)

# выводим результаты
print(f"Количество строк в исходном датафрейме: {initial_row_count}")
print(f"Количество строк в датафрейме после обработки: {final_row_count}")
print(f"Количество удаленных строк: {deleted_rows}")
print(f"Доля удаленных строк: {deleted_share}")
```
